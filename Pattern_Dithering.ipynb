```python
import cv2
import numpy as np

def pattern_dithering(image, block_size):
   

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    Modified_image = gray_img
    # Get the dimensions of the image
    height, width = gray_img.shape

    # Ensure the block size is either 2x2 or 3x3
    if block_size != 2 and block_size != 3:
        print("Block size must be either 2 or 3 for 2x2 or 3x3 sub-blocks.")
        return None

    # Calculate the number of blocks in rows and columns
    num_blocks_rows = height // block_size  # floor division 
    num_blocks_cols = width // block_size


    # Loop through the image and compute intensity for each sub-block
    for i in range(num_blocks_rows):
        for j in range(num_blocks_cols):
            # Define the boundaries of the sub-block
            start_row = i * block_size
            end_row = start_row + block_size
            start_col = j * block_size
            end_col = start_col + block_size

            # Get the sub-block
            sub_block = Modified_image[start_row:end_row, start_col:end_col]

            # Compute the average intensity of the sub-block
            average_intensity = np.mean(sub_block)
            # Compute the Decimal_value of the sub-block
            Decimal_value = average_intensity / 255.0
            
            if block_size == 2:
                counter = 1
                if 0 <= Decimal_value < 0.2:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            Modified_image[k,m] = 255.0
                elif 0.2 <= Decimal_value < 0.4:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 2:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0 
                            counter += 1 
                elif 0.4 <= Decimal_value < 0.6:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 2 or counter == 3:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0
                            counter += 1 
                elif 0.6 <= Decimal_value < 0.8:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 4:
                                Modified_image[k,m] = 255.0
                            else:
                                Modified_image[k,m] = 0.0  
                            counter += 1 
                elif 0.8 <= Decimal_value <= 1.0:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            Modified_image[k,m] = 0.0
            elif block_size == 3:
                counter = 1
                if 0 <= Decimal_value < 0.1:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            Modified_image[k,m] = 255.0
                elif 0.1 <= Decimal_value < 0.2:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 8:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0 
                            counter += 1      
                elif 0.2 <= Decimal_value < 0.3:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 8 or counter == 3:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0
                            counter += 1 
                elif 0.3 <= Decimal_value < 0.4:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 1 or counter == 8 or counter == 3:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0  
                            counter += 1 
                elif 0.4 <= Decimal_value <= 0.5:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 1 or counter == 8 or counter == 3 or counter == 7:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0 
                            counter += 1 
                elif 0.5 <= Decimal_value <= 0.6:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 1 or counter == 8 or counter == 3 or counter == 7 or counter == 7 or counter == 6:
                                Modified_image[k,m] = 0.0
                            else:
                                Modified_image[k,m] = 255.0  
                            counter += 1 
                elif 0.6 <= Decimal_value <= 0.7:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 2 or counter == 4 or counter == 5:
                                Modified_image[k,m] = 255.0
                            else:
                                Modified_image[k,m] = 0.0  
                            counter += 1 
                elif 0.7 <= Decimal_value <= 0.8:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 4 or counter == 5:
                                Modified_image[k,m] = 255.0
                            else:
                                Modified_image[k,m] = 0.0  
                            counter += 1 
                elif 0.8 <= Decimal_value <= 0.9:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            if counter == 5:
                                Modified_image[k,m] = 255.0
                            else:
                                Modified_image[k,m] = 0.0  
                            counter += 1 
                elif 0.9 <= Decimal_value <= 1.0:
                    for k in range(start_row,end_row):
                        for m in range(start_col,end_col):
                            Modified_image[k,m] = 0.0
    
    # Threshold to force pixel values to be either 0 or 255
    Modified_image = np.where(Modified_image >= 128, 255, 0).astype(np.uint8)
    return Modified_image

# Load an image
input_image = cv2.imread(r'E:\marg.jpg')
# Define a dither matrix ( 2x2 or 3x3)
Mask_Size = int(input("Enter the size of the mask: "))
# Apply pattern dithering using the defined mask
dithered_image = pattern_dithering(input_image, Mask_Size)
# Display or save the resulting image
cv2.imshow('Original Image', input_image)
cv2.imshow('Dithered Image', dithered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('Dithered3.jpg',dithered_image)

for row in dithered_image:
    print(row)
```
