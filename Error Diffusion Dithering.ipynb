```python
# Importing Libraries
import cv2
import numpy as np

# Read the image
image = cv2.imread(r'E:\marg.jpg')

# Convert the image to grayscale and convert data type from its current type to float32
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).astype(np.float32) #to avoide errors in adding and multiplication

# Define the Floyd-Steinberg dithering mask
mask = np.array([[0, 0, 7],
                 [3, 5, 1]]) / 16.0

# Get the height and width of the image
height, width = gray_image.shape

# Perform error diffusion dithering
for y in range(height - 1):
    for x in range(1, width - 1):
        # Fetch the pixel
        old_pixel = gray_image[y, x]
        new_pixel = 255 if old_pixel >= 128 else 0

        # Calculate the error
        error = old_pixel - new_pixel
        gray_image[y, x] = new_pixel

        # Error diffusion
        gray_image[y:y + 2, x - 1:x + 2] += mask * error

# Threshold to force pixel values to be either 0 or 255
gray_image = np.where(gray_image >= 128, 255, 0).astype(np.uint8)

# Displaying Original Image and Dithered Image
cv2.imshow('Original Image', image)
cv2.imshow('Dithered Image', gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('Dithered1.jpg',gray_image)

for row in gray_image:
    print(row)
```
