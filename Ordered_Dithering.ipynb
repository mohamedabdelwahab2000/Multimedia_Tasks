```python
import cv2
import numpy as np
def ordered_dithering(image, dither_matrix):
    # Convert the image to grayscale 
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    Modified_image = gray_image
    # Get the height and width of the image
    height, width = gray_image.shape
    mask_size = len(dither_matrix) # returns the number of rows in the array(dither_matrix)
    # Perform ordered dithering
    for y in range(height):
        for x in range(width):
            threshold = dither_matrix[y % mask_size][x % mask_size]
            if Modified_image[y, x] >= threshold:
                Modified_image[y, x] = 255.0
            else:
                Modified_image[y, x] = 0.0
    
    return Modified_image

# Load an image
input_image = cv2.imread(r'E:\marg.jpg')

# Define a dither matrix (ex 2x2 matrix)
n = int(input("Enter the size of the mask: "))
dither_matrix = []
print("Enter the mask elements:")
for i in range(n):
    row = []
    for j in range(n):
        element = float(input(f"Enter element for row {i+1}, column {j+1}: "))
        row.append(element)
    dither_matrix.append(row)

# Apply ordered dithering using the defined mask
dithered_image = ordered_dithering(input_image, dither_matrix)

# Display or save the resulting image
cv2.imshow('Original Image', input_image)
cv2.imshow('Dithered Image', dithered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('Dithered2.jpg',dithered_image)

for row in dithered_image:
    print(row)
```
